#### Question 1: For this challenge, forget how to add two numbers together. The best explanation on what to do for this function is this meme:



Examples:

i. meme_sum(26, 39) ➞ 515

2+3 = 5, 6+9 = 15

26 + 39 = 515

ii. meme_sum(122, 81) ➞ 1103

1+0 = 1, 2+8 = 10, 2+1 = 3

122 + 81 = 1103

iii. meme_sum(1222, 30277) ➞ 31499

#### Answer 1:

```python
def appendZero(n1,n2):
    
    if len(n1) == len(n2):
        pass
    elif len(n1) > len(n2):
        dif = len(n1) - len(n2)
        zero = '0'*dif
        n2 = zero+n2    
    else:
        dif = len(n2) - len(n1)
        zero = '0'*dif
        n1 = zero+n1
    return n1,n2   

def meme_sum(n1,n2):   
    n1, n2 = appendZero(str(n1),str(n2))
    cnt = 0
    i = 0
    sum = 0
    sumtotal = ''
    while cnt < len(n1):
        sum = int(n1[i]) + int(n2[i])
        sumtotal = sumtotal + str(sum)
        i += 1
        sum = 0
        cnt += 1
    return int(sumtotal) 

meme_sum(26, 39)
```

#### Question 2: Given an integer, create a function that returns the next prime. If the number is prime, return the number itself.

Examples:

i. next_prime(12) ➞ 13

ii. next_prime(24) ➞ 29

iii. next_prime(11) ➞ 11 (1 is a prime, so we return the number itself.)

#### Answer 2:

```python
def isPrime(n):
    if n==1:
        return True
    else:
        for i in range(2,int(n/2)+1):
            if n%i==0:
                return False
        return True

def next_prime(x):
    if isPrime(x):
        return x
    else:
        while(isPrime(x)!=True):
            x=x+1
        return x

next_prime(12)
```

#### Question 3: If a person traveled up a hill for 18mins at 20mph and then traveled back down the same path at 60mph then their average speed traveled was 30mph.

Write a function that returns the average speed traveled given an uphill time, uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate as an integer (mph). No rounding is necessary.

Examples

ave_spd(18, 20, 60) ➞ 30

ave_spd(30, 10, 30) ➞ 15

ave_spd(30, 8, 24) ➞ 12

#### Answer 3:

```python
def ave_spd(time1,speed1,speed2):

    uptime = time1/60
    #distance travelled uphill
    distance = speed1 * uptime
    #total distance
    totaldistance = distance*2
    #time taken to downhill
    timedown = distance/speed2
    totaltime = uptime + timedown
    totalspeed = totaldistance/totaltime
    return totalspeed

ave_spd(30, 8, 24)
```

#### Question 4: The Kempner Function, applied to a composite number, permits to find the smallest integer greater than zero whose factorial is exactly divided by the number.

kempner(6) ➞ 3

1! = 1 % 6 > 0
2! = 2 % 6 > 0
3! = 6 % 6 === 0

kempner(10) ➞ 5

1! = 1 % 10 > 0
2! = 2 % 10 > 0
3! = 6 % 10 > 0
4! = 24 % 10 > 0
5! = 120 % 10 === 0

A Kempner Function applied to a prime will always return the prime itself.

kempner(2) ➞ 2
kempner(5) ➞ 5

Given an integer n, implement a Kempner Function.

Examples:

i. kempner(6) ➞ 3

ii. kempner(10) ➞ 5

iii. kempner(2) ➞ 2

#### Answer 4:

```python
def isPrime(n):
    if n==1:
        return True
    else:
        for i in range(2,int(n/2)+1):
            if n%i==0:
                return False
        return True
def fact(num):
    factorial=1
    for i in range(1,num + 1):
        factorial = factorial*i
    #   print(factorial)
    return factorial
    
def kempner(n):
    if isPrime(n):
        return n
    else:
        for i in range(1,n):
            if fact(i)%n == 0:
                return i
            
kempner(10)
```

#### Question 5: You work in a factory, and your job is to take items from a conveyor belt and pack them into boxes. Each box can hold a maximum of 10 kgs. Given a list containing the weight (in kg) of each item, how many boxes would you need to pack all of the items?

Example:

i. boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2]) ➞ 5

Box 1 = [2, 1, 2, 5] (10kg)

Box 2 = [4, 3] (7kg)

Box 3 = [6, 1, 1] (8kg)

Box 4 = [9] (9kg)

Box 5 = [3, 2] (5kg)

#### Answer 5:

```python
def boxes(q):
    l = len(q)
    i = 0
    a = []
    box=[]
    while l :
        if sum(a)  <=  10 :
            a.append(q[i])
            i = i + 1
            l = l - 1   
        elif sum(a) > 10:
            box.append(a[:-1:])
            i = i - 1
            l = l + 1
            a.clear()
    box.append(a)

    print("We need {} boxes".format(len(box)))
    for i in range(len(box)):
        print("Box{} has {}, and {}kgs".format(i+1,box[i],sum(box[i])))
        
boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2])
```
