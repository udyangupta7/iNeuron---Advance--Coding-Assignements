#### Question 1: You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

Examples:

i. find_the_difference("abcd", "abcde") ➞ "e"

ii. find_the_difference("", "y") ➞ "y"

iii. find_the_difference("ae", "aea") ➞ "a"

#### Answer 1:

```python
def find_the_difference(s,t):
    sumS = 0
    sumT = 0
    for i in range(len(s)):
        sumS = sumS + (ord(s[i]))
    for j in range(len(t)):
        sumT = sumT + ord(t[j])
    dif = abs(sumS-sumT)
    return chr(dif)
find_the_difference('abcd', 'abcde')
```

#### Question 4: Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:

"nines:your answer, threes:your answer, ones:your answer"

You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).

Examples:

i. ones_threes_nines(10) ➞ "nines:1, threes:0, ones:1"

ii. ones_threes_nines(15) ➞ "nines:1, threes:2, ones:0"

iii. ones_threes_nines(22) ➞ "nines:2, threes:1, ones:1"

#### Answer 4:

```python
def ones_threes_nines(num):
    dict = {'nines':0,'threes' : 0,'ones':0 }
    strDict = ''
    while(num > 0):
        if num // 9 > 0:
            dict['nines'] = num // 9
            num = num % 9
        elif num // 3 > 0:
            dict['threes'] = num // 3
            num = num % 3
        else:
            dict['ones'] = num // 1
            num = num % 1
    for k,v in dict.items():
        strDict = strDict + k + ":" + str(v) + " "
    return ', '.join(strDict.strip().split())

ones_threes_nines(22)
```
